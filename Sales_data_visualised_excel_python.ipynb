```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
```

### <font color=black> Taking out useful data, ordering and light processing </font>

```python
data = pd.read_csv("sales_data_sample.csv", sep=",", encoding='Latin-1')
#print(list(data.columns))
data['Sales_person'] = data.apply(lambda row: row['CONTACTFIRSTNAME'] + ' ' + row['CONTACTLASTNAME'], axis=1)
useful_data = data[['ORDERDATE', 'ORDERNUMBER' , 'CUSTOMERNAME' ,'Sales_person' ,'COUNTRY' ,'PRODUCTLINE' ,'PRICEEACH' ,'QUANTITYORDERED' ,'SALES','QTR_ID', 'MONTH_ID', 'YEAR_ID']]
```

```python
useful_data.columns = ['Date', 'Customer_ID', 'Customer_Name', 'Sales_Person', 'Country', 'Item', 'Price', 'Quantity', 'Revenue','Quarter_ID','Month_ID', 'Year_ID']
```

### Create appropriate date-time format to be used in MS Excel

```python
useful_data['Date'] = pd.to_datetime(useful_data['Date']).dt.date
useful_data.head(5)
```

### Output CSV file to be further processed and visualised in MS Excel

```python
useful_data.to_csv("output.csv")
```
